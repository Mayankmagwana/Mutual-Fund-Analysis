```python
import requests
import pandas as pd
import numpy as np
import os
```

```python
Date = "January-2013"
```

Importing file from HDFC MUTUAL FUNDS website

```python
url = "https://files.hdfcfund.com/s3fs-public/2020-04/Monthly Portfolios for March 2020_1.xls"

response = requests.get(url)

filename = "hdfcMarch2020.csv"

with open(filename,"wb") as file1:
    for row in response:
        file1.write(row)
if response.status_code == 200:        
    print("File is imported")
```

```python
filename = "HDFC_monthly_portfolio/May 2020.xls"
```

Cleaning Mutual Fund Data

```python
fund = "HDFCEQ"

df = pd.read_excel(filename, sheet_name = fund, skiprows = 4,usecols = [1,2,3,4,5,6,7])
df.head()

df.columns = ["ISIN","Coupon","instrument","industry_rating","qty","marketval","pct"]
df["Date"] = Date
df.head()
```

```python
for value in df.ISIN:
    if "Grand Total" in str(value):
        total = df.loc[df['ISIN'] == value]['marketval']
print(total)
```

```python
df['Coupon'].fillna(0, inplace = True)


#Finding Net Asset
#value = df.loc[df['ISIN'] == 'Grand Total']['marketval']
for value in df.ISIN:
    if "Grand Total" in str(value):
        total = df.loc[df['ISIN'] == value]['marketval']
actual_total = float(total)
print('Actual ToTal is :',actual_total)


#TREPS - Tri-party Repo
#Net Current Assets
#Collateralised Borrowing & Lending Obligation
#Net Current Assets
#HDFC Liquid Fund - Direct Plan - Growth Plan
#HDFC Liquid Fund - Direct Plan - Growth Plan 

#Kridhan Infra Ltd.
#NCL Industries Ltd.
#State Bank of Indore**
#Gold Bar
#Cash Management Savings - Direct Plan - Growth Option
#Cognizant Technology Solutions Corp


df.loc[(df.instrument == 'Net Current Assets'),['Coupon','industry_rating','ISIN','qty']] = [0,'Net Current Assets','NCA',1]
df.loc[(df.instrument == 'Gold - Gold'),['Coupon','industry_rating','ISIN','qty']] = [0,'Gold','G-G',1]
df.loc[(df.instrument == 'NIFTY NSE PUT OPTION STRIKE PRICE 9600 EXPIR JUN 2020'),['ISIN','industry_rating']] = ['NIFTY PUT 9600','NIFTY Option']
df.loc[(df.instrument == 'NIFTY NSE PUT OPTION STRIKE PRICE 9900 EXPIR JUN 2020'),['ISIN','industry_rating']] = ['NIFTY PUT 9900','NIFTY Option']
df.loc[(df.instrument == 'HDFC Corporate Bond Fund - Growth Option - Direct Plan'),['ISIN','industry_rating']] = ['HDFC Cop Bond','Fund']

df.loc[(df.ISIN == 'Collateralised Borrowing & Lending Obligation'),['Coupon','industry_rating','instrument','qty']] = [0,'Borrowing and Lending','Collateralised Borrowing & Lending Obligation',1]
df.loc[(df.ISIN == 'Net Current Assets'),['Coupon','industry_rating','instrument','qty']] = ['0','Net Current Asset','additional2',1]
df.loc[(df.instrument == 'HDFC Liquid Fund - Direct Plan - Growth Plan'),['industry_rating','qty']] = ['Fund',1]
df.loc[(df.instrument == 'Cash Management Savings - Direct Plan - Growth Option'),['industry_rating','qty']] = ['Fund',1]
df.loc[(df.instrument == 'Cognizant Technology Solutions Corp'),['ISIN','industry_rating']] = ["US1924461023",'OverSea Bond']



for value in df.instrument: 
    if 'Option' in str(value):
        df.loc[(df.instrument == value),['Coupon','industry_rating']] = [0,'Fund']
    
    if 'Direct Plan' in str(value):
        df.loc[(df.instrument == value),['Coupon','industry_rating']] = [0,'Fund']
    
    if 'Collateralised Borrowing & Lending Obligation' in str(value):
        df.loc[(df.instrument == value),['ISIN','Coupon','industry_rating','qty']] = ['Collateralised Borrowing & Lending Obligation',0,'Borrowing and Lending',1]    
    
    if 'Traded Fund' in str(value):
        df.loc[(df.instrument == value),['Coupon','industry_rating']] = [0,'Traded Fund']
                    
    if 'Gold Bar' in str(value):
        df.loc[(df.instrument == value),['industry_rating']] = ['Gold']
        
    if 'Safari Industries (India) Ltd' in str(value):
        df.loc[(df.instrument == value),['industry_rating']] = ['Corporate Bond']
    
    
    if 'Repo' in str(value):
        df.loc[(df.instrument == value),['Coupon','industry_rating','ISIN','qty']] = ['0','Repo','Repo',1]
    
    if 'Ltd.' or 'Limited' or 'Bank' in str(value):
        isin = list(df.loc[(df.instrument == value)]['ISIN'])
        try:
            if isin[-1] is np.nan:
                df.loc[(df.instrument == value),['Coupon','industry_rating','ISIN','qty']] = ['0','additional','additional',1]
        except IndexError:
            continue
    else:
        continue





    

#print(df.isnull())
df.dropna(inplace = True)
df.reset_index(drop = True, inplace = True)


for value in df.industry_rating:
    value = str(value)
    if value[-1] == 'A' or value[-1] == "-" or value[-1] == "+" or value[-2] == "O":
        df.loc[(df.industry_rating == value),["industry_rating"]] = ["Corporate Bond"]
    else:
        continue

#calculated total

#for value in df.marketval:
  
Index_label = df[df['marketval'] == 'Nil'].index.tolist()
df.drop(df.index[Index_label],inplace = True)
 
    
    
    
calculated_total = df['marketval'].sum()
print("Calculated Total : ", calculated_total)

diff = actual_total-calculated_total
if abs(diff) < 1:
    print("Tallied")
else:
    print("There is a differenc of : ",actual_total - calculated_total )

if not os.path.exists('funds_dfs'):
    os.makedirs('funds_dfs')
    
df.to_csv('funds_dfs/{}.csv'.format(fund))
print('CSV Published')
```
